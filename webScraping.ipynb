```python
from bs4 import BeautifulSoup
import requests
import re

searchTerm=input("Enter the product that you want to search for: ")
url = f"https://mdcomputers.in/index.php?search={searchTerm}&submit_search=&route=product%2Fsearch"
res=requests.get(url).text;
doc=BeautifulSoup(res,'html.parser')

page_text = doc.find(class_="col-sm-6 text-right")
```

```python
pages=page_text.text.split()[-2][1:]
itemsFound={}

for page in range(1,int(pages)+1):
  url=f"https://mdcomputers.in/index.php?route=product/search&page={page}&search={searchTerm}"
  page=requests.get(url).text;
  doc=BeautifulSoup(page,"lxml")
  div=doc.find(class_="products-category")
  items=div.find_all('h4')

  for item in items:
    parent=item.parent
    link=item.a['href'];
    new_price=parent.find(class_="price-new").text.replace(',','')[1:]
    itemsFound[item.text]={"price":new_price,"link":link};
```

```python
sorted_items=sorted(itemsFound.items(),key= lambda x: int(x[1]['price']));
for item in sorted_items:
  print(item[0]);
  print(f'Rs{item[1]["price"]}')
  print(item[1]["link"])
  print(".................................................................................")
```
